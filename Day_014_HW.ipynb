```python
# 載入基礎套件
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### 負相關

```python
# 隨機生成 1000 個介於 0~50 的數 x
x = np.random.randint(0, 50, 1000)

# 加一些雜訊
y =  -x + np.random.normal(0, 5, 1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x, y)
```

```python
# 負相關的 x,y 分布
plt.scatter(x, y)
```
