Q1. LASSO 回歸可以被用來作為 Feature selection 的工具，請了解 LASSO 模型為什麼可用來作 Feature selection

> Ans:<br>
>   參考[資料1](https://www.youtube.com/watch?v=NGf0voTMlcs&t=9s) 及 [資料2](https://www.youtube.com/watch?v=Xm2C_gTAl8c) <br>
>  <br>
>   Lasso regression penalty 在 λ 較大的情況下, 偏差(sum of square residual + λ*| slope |)最小值會在 slope=0 <br>
>   此時, 對於目標變數的預測, 可忽略該變數的影響, 故在複迴歸時, 可用來篩選特徵



Q2. 當自變數 (X) 存在高度共線性時，Ridge Regression 可以處理這樣的問題嗎?

> Ans:<br>
>   可以, 藉由L2 penalty調整權重，減少特徵共線性所造成的Over-fitting。<br>
